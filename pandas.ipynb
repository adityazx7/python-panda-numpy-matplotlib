```python
import pandas as pd
# pandas means panel data
# excel type

# series is pandas 1 dimensional labelled array

data = [1000, 102,104]
series = pd.Series(data)

print(series)
```

```python
data = [100.1, 102,104]
series = pd.Series(data)

print(series)
```

```python
data = ['A', 'B', 'C']
series = pd.Series(data)

print(series)
```

```python
data = [True, False, True]
series = pd.Series(data)

print(series)
```

```python
data = [1000, 102,104]
series = pd.Series(data, index = ["a", "b", "c"])

print(series)
series.loc["c"] = 200
print(series.loc["c"])
```

```python
data = [1000, 102,104]
series = pd.Series(data, index = ["a", "b", "c"])

print(series.iloc[1])
```

```python
data = [1000, 102,104, 200, 202, 204]
series = pd.Series(data, index = ["a", "b", "c", "d","e", "f"])

print(series[series < 200])
```

```python
calories = {"Day 1": 1750, "Day 2": 2100, "Day 3": 1700}
series = pd.Series(calories)

print(calories)

series.loc["Day 3"] += 500

print(series.loc["Day 3"])

print(series[series > 2000])
```

```python
import pandas as pd
# Data Frame

data = {"Name":["Spongebob", "patrick", "aditya"],"Age": [30, 35, 18]}
df = pd.DataFrame(data, index = [1, 2, 3])
print(df)

print(df.loc[3])

print(df.iloc[1])

# add a new column
df["job"] = ["Cook", "N/A", "dev"]
print(df)


# add a new row
new_row = pd.DataFrame([{"Name": "Sandy", "Age": 28, "job": "engineer"}, {"Name": "Randy", "Age": 68, "job": "brand"}], index = [4, 5])
df = pd.concat([df, new_row])
print(df)
```
